```python
#question 1
import plotly
import plotly.graph_objects as go
import plotly.express as px

np.random.seed(30)
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}
df = pd.DataFrame(data)
df
```

```python
fig=go.Figure()
fig.add_trace(go.Scatter3d(x=df.X,y=df.Y,z=df.Z, mode='markers',marker_size=2))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
import plotly
import plotly.graph_objects as go
import plotly.express as px
```

```python
#question 4
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

fig= go.Figure()
fig.add_trace(go.Scatter3d(x=df.X,y=df.Y,z=df.Z, mode='markers'))
```

```python
#question 5
np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK',
'Germany', 'France'],
    'Population':
np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000,
5)
}
df = pd.DataFrame(data)

fig=go.Figure()
fig.add_trace(go.Scatter(x=df.GDP,y=df.Population,marker_size=0.1*df['Population']))

```

Bokeh

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
from bokeh.plotting import figure,show

x=np.random.randint(0,10,100)
y=np.sin(x)
p= figure(title="sine wave",x_axis_label='x',y_axis_label='y')
p.line(x,y,line_width=2)
show(p)
```

```python
#question 2
x= np.random.rand(100)
y=np.random.rand(100)
size = np.random.randint(1,10,100)
color=np.random.choice(['green','blue','yellow','pink'])
data=pd.DataFrame({'x':x,'y':y,'size':size,'color':color})
data
```

```python
from bokeh.model import ColumnDataSource 
source=ColumnDataSource(data)
p=figure(title="hi",y_axis_label='y',x_axis_label='x')
p.scatter('x','y',size='size',color='color',source='source')
show(p)
```
